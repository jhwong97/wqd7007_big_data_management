```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MaxNLocator
from matplotlib.ticker import MultipleLocator
```

```python
# Generate 10 random numbers
list1 = random.sample(range(0, 30), 10)
repeated_numbers = list1 * 10
```

```python
random_numbers = [random.randint(0, 30) for i in range(100)]
```

```python
# Count the frequency of each element
frequency_dict_repeated = {}
for item in repeated_numbers:
    if item in frequency_dict_repeated:
        frequency_dict_repeated[item] += 1
    else:
        frequency_dict_repeated[item] = 1
```

```python
# Count the frequency of each element
frequency_dict_random = {}
for item in random_numbers:
    if item in frequency_dict_random:
        frequency_dict_random[item] += 1
    else:
        frequency_dict_random[item] = 1
```

```python
# Histogram for repeated numbers
numbers_A = list(frequency_dict_repeated.keys())
frequencies_A = list(frequency_dict_repeated.values())

# Create a histogram
plt.bar(numbers_A, frequencies_A)

# x-axis and y-axis formatting
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(2))
ax.yaxis.set_major_locator(MaxNLocator(integer=True))

# Add labels and title
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.title("Histogram A")

# Show the histogram
plt.show()
```

```python
data_repeated = {'Numbers' : numbers_A,
        'Frequency' : frequencies_A}
pd.DataFrame(data_repeated)
```

```python
# Histogram for random numbers
numbers_B = list(frequency_dict_random.keys())
frequencies_B = list(frequency_dict_random.values())

# Create a histogram
plt.bar(numbers_B, frequencies_B)

# x-axis and y-axis formatting
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(2))
ax.yaxis.set_major_locator(MaxNLocator(integer=True))

# Add labels and title
plt.xlabel("Numbers")
plt.ylabel("Frequency")
plt.title("Histogram B")

# Show the histogram
plt.show()
```

```python
data_random = {'Numbers' : numbers_B,
               'Frequency' : frequencies_B}
pd.DataFrame(data_random)
```
